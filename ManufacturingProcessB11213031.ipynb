```python
!pip install python-barcode[images] qrcode[pil]
import barcode
from barcode.writer import ImageWriter
from IPython.display import Image, display

# 產品料號
product_id = input("bohan製程條碼展示-請輸入你設計的物料編號")

# [很重要]建立 Code128 條碼（未來實務專題可以把 code128 改成廠商類型）
code = barcode.get('code128', product_id, writer=ImageWriter())

# 儲存條碼
filename = code.save("product_barcode")

# 顯示條碼
display(Image(filename=filename))
```
